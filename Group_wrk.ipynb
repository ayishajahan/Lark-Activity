```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Agricultrue_labour_wages.csv")
```

# Exploratory Data Analysis

UNIVARIATE ANALYSIS AND BIVARIATE ANALYSIS

```python
data.hist(figsize=(10,10))
plt.show()
```

```python
count=data["Labour Category"].value_counts()
count.to_frame()
plt.figure(figsize=(10,10))
plt.pie(count,labels=count.index,autopct='%1.2f%%')
plt.title("Distribution of the Labour Category in company ")
plt.show()
```

```python
columns = ["Centre", "Gender", "Labour Category", "Labour Type", "State", "District",
          "Wage(INR)", "Month", "Year", "Wage (Dollar)"]
```

```python
for i in columns:
    plt.figure(figsize=(15, 5))
    sns.countplot(x=data[i])
    print("Distribution of ",i )
    plt.show()
```
