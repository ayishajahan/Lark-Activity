```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Question 1

```python
zen=pd.read_csv('telco.csv')
```

```python
zen.head()
```

```python
sns.countplot(x='gender', hue='Churn', data=zen)
plt.title("Churn count vs Gender")
plt.grid()
```

# Question 2

```python
SCF=zen.loc[(zen["gender"]=="Female") & (zen["SeniorCitizen"]==1)]
SCF
```

```python
sns.countplot(SCF["SeniorCitizen"])
plt.xticks
plt.title("Senior Female Citizens")
plt.grid()
```

# Question 3

```python
plt.scatter(zen["tenure"],zen["TotalCharges"],s=10,marker="*")
plt.xlabel("Tenure")
plt.ylabel("Total Charges")
plt.title("Relation between tenure vs Total charges")
```

# Question 4

```python
SC=zen.loc[(zen["SeniorCitizen"]==1)]
SC.head()
```

```python
sns.countplot(SC["SeniorCitizen"],hue=SC["Contract"])
plt.xticks(visible = False)
plt.grid()
```

# Question 5

```python
sns.countplot(zen["PaymentMethod"],hue=zen["gender"])
plt.xticks(rotation=45)
plt.title("Prefered payment method Vs Gender")
plt.grid()
```

```python
sns.countplot(zen["PaymentMethod"],hue=zen["Churn"])
plt.xticks(rotation=45)
plt.title("Prefered payment method Vs Churn")
plt.grid()
```

```python
sns.countplot(zen["PaymentMethod"],hue=zen["SeniorCitizen"])
plt.xticks(rotation=45)
plt.title("Prefered payment method Vs SeniorCitizen")
plt.grid()
```

```python
sns.countplot(zen["PaymentMethod"],hue=zen["Contract"])
plt.xticks(rotation=45)
plt.title("Prefered payment method Vs Contract")
plt.grid()
```
