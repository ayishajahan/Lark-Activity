# 5. Check and handle outliers in at least 3 columns in the dataset

```python
sns.boxplot(data=zen[['Age', 'Fare', 'SibSp']], orient='h')
plt.show()
```

```python
zen_with_outliers = zen.copy(deep=True)
```

```python
def outlier_handling(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    q_limit = 1.5 * iqr
    upper_limit = q3 + q_limit
    lower_limt = q1 - q_limit
    outliers = data.loc[(data < lower_limt) ^ (data > upper_limit)].tolist()
    idx = data.loc[(data < lower_limt) ^ (data > upper_limit)].index.tolist()
    return outliers, idx
```

```python
fare_outlier_list, fare_outlier_index = outlier_handling(zen['Fare'])
age_outlier_list, age_outlier_index = outlier_handling(zen['Age'])
sibsp_oulliers, sibsp_outlier_index = outlier_handling(zen['SibSp'])
```

```python
print(len(fare_outlier_index), len(age_outlier_index), len(sibsp_outlier_index))
```

```python
new_idx = sibsp_outlier_index + age_outlier_index + fare_outlier_index
len(new_idx)
```

```python
new_idx = set(new_idx)
new_idx
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
lr=linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
predictions=model.predict(x_test)
print('MSE is : ',mean_squared_error(y_test,predictions))
print('r squared value is : ',r2_score(y_test,predictions))
```
